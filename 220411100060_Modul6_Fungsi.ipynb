2.Jenis-jenis Fungsi dan perbedaannya yaitu:
>
- **Fungsi tanpa parameter/argumen**
adalah bentuk sederhana dari rumus fungsi yang berarti bahwa fungsi tersebut dapat dieksekusi cukup dengan memanggil fungsinya saja tanpa perlu adanya argumen tertentu.dan juga fungsi tersebut tidak akan mengembalikan nilai apapun kepada perintah pemanggilnya. contoh syntax:
```
def nama_fungsi():
    syntax1 / print(txt)
    ...
```
- **Fungsi dengan Argument/Parameter**
adalah kumpulan perintah yang dikelompokkan menjadi satu dengan suatu variabel yang berfungsi untuk menampung nilai yang akan dikirimkan ke dalam fungsi.itulah yang disebut dengan fungsi tanpa parameter contoh syntax:
```
def nama_fungsi(parameter,..):
    syntax1 / print(parameter)
```
- **Fungsi tanpa return value**
adalah fungsi yang digunakan untuk melakukan proses-proses yang tidak menghasilkan value, seperti halnya melakukan looping, proses yang melibatkan pengubahan nilai ataupun yang lainnya.contoh syntax:
```
def nama_fungsi(parameter,..):
     syntax1
      ....
      syntax n
```
- **Fungsi dengan return value**
adalah fungsi yang memiliki nilai balik yang bertujuan agar dapat mengembalikkan nilai yang bisa digunakan untuk operasi-operasi lainya pada code,syntax umumnya yaitu:
```
def nama_fungsi(parameter,.):
     syntax1
      ....
      syntax n
    return value
```

##### 2.1 Cek List

Buatlah Fungsi dan code untuk memanggil fungsi dengan menggunakan Python, antara lain :

• Fungsi createList untuk membentuk suatu list dengan parameter atau argumen berupa ukuran list
>
• Fungsi isPrime untuk menentukan apakah suatu bilangan adalah bilangan prima, dengan argumen
adalah bilangan yang akan dicek
>
• Fungsi createListPrime untuk membuat suatu list yang berisi bilangan prima, dengan ketentuan :
>
– argumen atau paramater adalah list dimana di setiap anggotanya akan dicek termasuk bilangan
prima ataukah tidak
>
– gunakan fungsi isPrime yang sudah dibuat, untuk cek apakah suatu bilangan adalah bilangan
prima
>
– anggota-anggota list yang dihasilkan dari fungsi ini, tidak boleh sama [Hint: gunakan in]
>
Contoh output yang dihasilkan dapat dilihat pada Gambar 1.
>
image.png
>

>

>
Size List utama =15
 

````python
def createList (ukuranlist):
    list1=[]
    for i in range(ukuranlist):
        inputan=int(input(f'Masukkan data ke-{i}:'))
        list1.append(inputan)
    return list1
def isPrime(bilangan):
    count=0
    for i in range(1,bilangan+1):
        if bilangan%i==0:
            count=count+1
    if count==2:
        return True
    return False
def createListPrime(a):
    create=[]
    for i in a :
        if isPrime(i) and i not in create:
            create.append(i)
    return create
data=createList(15)
print('data list =',data)
prima=createListPrime(data)
print('data prima = ',prima)
````
