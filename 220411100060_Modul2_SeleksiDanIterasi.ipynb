##### 2.1 Bilangan Prima
Buatlah flowchart (tidak harus menggunakan flogorithm) dari penentuan apakah suatu bilangan (inputan
dari user) adalah bilangan prima. Implementasikan flowchart yang sudah dibuat (bukan generate dari
flowgorithm) dengan menggunakan bahasa Python, sehingga menghasilkan output seperti contoh-contoh
yang terdapat pada Gambar 1.

### jawab:
image.png<br>


```python
x=int(input('masukkan bilangan ='))
counter=0
pembagi=" "
for i in range(1,x+1):
    sisa=x%i
    if sisa==0:
        pembagi+"  "+str(i)
        counter+=1
if counter==2:
    print(x," adalah bilangan prima")
else:
    print(x,'bukan prima',',',"memiliki jumlah faktor pembagi ",counter)
```

```python
a=int(input('Panjang sisi pertama dari segitiga:='))
b=int(input('Panjang sisi kedua dari segitiga='))
c=int(input('Panjang sisi ketiga dari segitiga='))
if c*c==a*a+b*b:
    print("segitiga siku-siku dengan sisi siku-siku adalah",a,"dan",b,"serta sisi miring adalah",c)
else:
    print("sisi","{",a,b,c,"}","Bukan segitiga siku-siku" )
```

##### 2.3 Iterasi di dalam Iterasi
Buatlah code untuk menampilkan penomoran daftar isi, seperti contoh pada 4. Dengan inputan dari user
berupa jumlah bab dan sub bab dalam setiap bab. [Hint : Buatlah iterasi di dalam iterasi]
#### jawab:

```python
a=int(input("masukkan jumlah bab="))
b=int(input("masukkan jumlah sub bab di setiap bab="))
for i in range(a):
    print(i+1,".")
    for ch in range(1,b+1):
        x=i+1
        print("    ",x,".",ch)
```

##### 2.4 Menu
Buatlah code untuk membuat menu, dimana setiap menu melakukan operasi yang berbeda :
Menu :
• Tekan 1 untuk operasi perhitungan luas lingkaran (input adalah jari-jari)
• Tekan 2 untuk operasi perhitungan luas persegi panjang (input adalah panjang dan lebar)
• Tekan 3 untuk operasi perhitungan luas segitiga (input adalah alas dan tinggi)
Tambahkan pilihan bagi user, apakah ingin mengulangi operasi kembali ataukah tidak, jika ‘y’, maka tampilkan menu kembali dan lakukan operasi sesuai pilihan dari user. Jika user memilih ‘t’ maka operasi berhenti. Contoh output dari code tersebut dapat dilihat pada Gambar 5
#### jawab:

```python
stop=False
while not (stop):
    print("Menu")
    print('Tekan 1 untuk operasi perhitungaan luas lingakaran(input adalah jari-jari)')
    print("Tekan 2 untuk operasi perhitungan luas persegi panjang (input adalah panjang dan lebar)")
    print("Tekan 3 untuk perhitungan perhitungan luas segitiga(input adalah alas dan tinggi)")
    a=int(input('masukkan pilihan anda='))
    if a==1:
        r=int(input("masukkan jari-jari="))
        L=22/7*r*r
        print("Luas lingkaran=",L)
    if a==2:
        p=int(input("masukkan panjang="))
        t=int(input("masukkan tinggi="))
        LP=p*t
        print("Luas persegi panjang=",LP)
    elif a==3:
        alas=int(input("masukkan alas="))
        tinggi=int(input("masukkan tinggi="))
        LS=(alas*tinggi)/2
        print("Luas segitiga=",LS)
    pilihan=input("ingin mengulangi operasi kembali (y/t)?")
    if pilihan=="y":
        stop=False
    else:
        stop=True
    print(43*'==')
```
